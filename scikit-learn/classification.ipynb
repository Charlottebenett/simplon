```python
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
```

```python
data = pd.read_csv('logement-encadrement-des-loyers (1).csv',sep = ';')
data.head()
```

```python
X = pd.DataFrame(data,columns = ['annee','piece',
                                 'epoque','id_zone'])
Y = data['ref']

X = pd.get_dummies(X)

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, 
                                          test_size = 0.2)
```

```python
regressor = RandomForestRegressor(n_estimators = 10, 
                                  random_state = 0)

regressor.fit(X_train, Y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
import matplotlib.pyplot as plt

plt.title('Performance du modèle')
plt.plot(list(y_pred)[:20], label='prediction client')
plt.plot(list(Y_test)[:20], label='estimation-ml')
plt.legend()
```

```python
import pandas as pd

df = pd.read_csv('reduced_covid_sexe.csv')
df = df.iloc[2:]

df.head(15)
```

```python
# créer une colomne season et l' ajouter une colomne season au dataframe 

df = df.iloc[2:]
dates = pd.to_datetime(df['Unnamed: 0'])

#définir les quatres saisons 
def get_season(dates):
    year = str(dates.year)
    seasons = {'spring': pd.date_range(start='21/03/'+year, end='20/06/'+year),
               'summer': pd.date_range(start='21/06/'+year, end='20/09/'+year),
               'autumn': pd.date_range(start='21/09/'+year, end='20/12/'+year)}
    if dates in seasons['spring']:
        return 'spring'
    if dates in seasons['summer']:
        return 'summer'
    if dates in seasons['autumn']:
        return 'autumn'
    else:
        return 'winter'

#ajouter la colone season au dataframe
df['season'] = dates.map(get_season)
 
df
```

```python
#supprimer colomne unnamed: 0

df = df.drop(columns=['Unnamed: 0'])

#décomposer la colomne season en 4 colomne spring summer autumn winter 
df_2 = df.copy() #on copie le dataframe pour ne pas modifier l'original
df_2 = pd.get_dummies(df_2, columns=['season'])
```

```python
# Afficher les valeurs nulles dans le DataFrame
df.isnull().sum()
```

```python
X = pd.DataFrame(df,columns = ['hosp.1','hosp.2',
                                 'rea.1','rea.2'])
Y = df['dc']



X_train, X_test, Y_train, Y_test = train_test_split(X, Y,  test_size = 0.2)
```

```python
regressor = RandomForestRegressor(n_estimators = 10, 
                                  random_state = 0)

regressor.fit(X_train, Y_train)
```

```python
#calculer le score de mon model 
regressor.score(X_test, Y_test)
```

```python
y_pred = regressor.predict(X_test)
```

```python
import matplotlib.pyplot as plt

plt.title('Performance du modèle')
plt.plot(list(y_pred)[:30], label='prediction dc')
plt.plot(list(Y_test)[:30], label='estimation-ml')
plt.legend()
```

```python
#plot le mean squared error 
from sklearn.metrics import mean_squared_error
mean_squared_error(Y_test, y_pred)
mean_squared_error
#plot the histogram of residuals
import seaborn as sns
sns.distplot(Y_test - y_pred)
plt.show()

#plot the scatter plot of residuals
plt.scatter(Y_test, Y_test - y_pred, c = 'blue')
plt.hlines(y = 0, xmin = 0, xmax = 50, linewidth = 2)
plt.title('Residual plot')
plt.show()

```
